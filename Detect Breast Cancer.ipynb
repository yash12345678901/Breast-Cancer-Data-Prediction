```python
# split the data set independent(X) and dependent(Y) dataset
X = df.iloc[:,2:31].values
Y = df.iloc[:,1].values
type(df)
```

```python
# split the dattaset 75% training and 25% testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
# scale the data (feature scalling)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

```python
X_test.shape
```

```python
# create the functions of the models
def models(X_train, Y_train):
    
    # Logistic Regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state = 0)
    log.fit(X_train, Y_train)
    
    # Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    tree.fit(X_train, Y_train)
    
    # Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
    forest.fit(X_train, Y_train)
    
    # Print the model Accuracy on the training data
    print('[0] Logistic Regression Training Accuracy:', log.score(X_train, Y_train))
    print('[1] Decision Tree Classifier Training Accuracy:', tree.score(X_train, Y_train))
    print('[2] Random Forest Classifier Training Accuracy:', forest.score(X_train, Y_train))
    
    
    return log, tree, forest
    
    
```

```python
# Getting all of the values
model = models(X_train, Y_train)
```

```python
# test model accuracy on test data on confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, model[0].predict(X_test))
```

```python
for i in range(len(model)):
    print('Model', i)
    TP = cm[0][0]
    TN = cm[1][1]
    FN = cm[1][0]
    FP = cm[0][1]
    print(cm)
    print("Testing Accuracy :" , (TP + TN)/(TP + TN + FP + FN))
```

```python
# another way to get the metrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
for i in range(len(model)):
    print('Model', i)
    print(classification_report(Y_test, model[0].predict(X_test)))
    print(accuracy_score(Y_test, model[0].predict(X_test)))
```
